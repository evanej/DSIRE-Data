```python
# This notebook will take a list of searchterms and produce a dictionary with the
#     number of times they occur in a pandas dataframe and search through that
#     dataframe and produce files for where those terms exist
```

```python
# First, import the data as a pandas dataframe
```

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv(r'/Users/Jax/Desktop/Research_with_Evan/pgadmin_data_tables/pgadmin_dsire_merge_1.csv')
#data
```

```python
#Make a list of all the headers in the data
headers = []
for col in data.columns:
    headers.append(col)
#headers
#len(headers)
```

```python
# We will a list of unique cells from the data and convert it into a long string
#    so that we can accurately count how often specific words appear in the data:
```

```python
df_string = data.applymap(str)
```

```python
# Convert each column into a string of it's unique values
#   and combine all of those strings into one list:
```

```python
data_string = []

for col in headers:
    a_col = df_string[col]
    
    a_col_list = a_col.tolist()
    
    unique_in_col = []
    
    for value in a_col_list:
        if value not in unique_in_col:
            unique_in_col.append(value)
    unique_col_string = " ".join(unique_in_col)
    unique_col_string = unique_col_string.lower()
    data_string.append(unique_col_string)
```

```python
# Returns a dictionary with the number of times a list of search
#    terms appears in all of the data:
```

```python
searchterm_counts = {}

# the data has been converted to all lower case, so only include
#    lower case searchterms
searchterms = ['cost share','cost sharing','disability', 'disabled',
               'disadvantage','diversity','entrepreneur','equal',
               'equity','fund share','high tech',
               'high-tech','in need','innovation','innovative',
               'low income','low-income', 'matching fund', 'matching grant',
               'minority', 'payback',
               'pay back','poor',
               'poverty','race','rebate','rural','small business',
               'small firm','sole proprietor','state match',
               'underserved','underserved population',
               'victim','vulnerable','women']


for word in searchterms:
    col_count = []
    for num in range(len(data_string)):
        col_count.append(data_string[num].count(word))
    total_count = 0
    total_count = sum(col_count)
    
    searchterm_counts[word] = total_count
    
searchterm_counts

```

```python
# To export the search term dictionary as an Excel document:
```

```python
searchterm_counts_df = pd.DataFrame(data = searchterm_counts, index=['search term counts'])
searchterm_counts_df = (searchterm_counts_df.T)
print(searchterm_counts_df)
searchterm_counts_df.to_excel('searchterm_counts.xlsx')
```
